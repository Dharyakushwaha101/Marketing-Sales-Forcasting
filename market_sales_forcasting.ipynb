```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
market_dataset = pd.read_csv("/content/Walmart.csv")
```

```python
market_dataset.head()
```

```python
market_dataset.shape
```

```python
X = market_dataset.drop(['Weekly_Sales','Date'],axis=1)
print(X)
```

```python
Y = market_dataset['Weekly_Sales']
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

```python
#model evaluation
#prediction on training data
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train,training_data_prediction)
print("R squared error : ",error_score)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Sales atrtribute')
plt.ylabel("Weekly_Sales")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

```python
# prediction on test data
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_test,test_data_prediction)
print("R squred error:",error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel('Sales atrtribute')
plt.ylabel("Weekly_Sales")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
